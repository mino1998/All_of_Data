## TV시사교양

```python
from selenium import webdriver
from selenium.webdriver.support.ui import Select
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from urllib.parse import urljoin
# 크롬드라이버를 설치한 경로 chromedriver.exe가 존재하는 path를 지정합니다.
webdriver_path = 'chromedriver'

# Chrome WebDriver 실행
browser = webdriver.Chrome(webdriver_path)

url = "https://www.themoviedb.org/"

browser.implicitly_wait(3)

browser.get(url)  # 페이지 접속

name_ls=[]

plot_ls=[]

button_ls =[]

genre_ls=[]

img_Is=[]

from selenium.common.exceptions import NoSuchElementException

length=len(vod_list)

for i in range(0, length):
    name='none'
    plot='none'
    genre='none'
    img='none'
    button='none'
    browser.implicitly_wait(3)
    browser.get(url)  # 페이지 접속
    search_input = browser.find_element(By.XPATH, '/html/body/div[1]/main/section[1]/div/div/div/div[2]/form/label/input')
    search_input.send_keys(vod_list[i])
    search_input.send_keys(Keys.RETURN)
    browser.implicitly_wait(3)  # 대기시간 부여하는 코드
    # 검색 결과가 없는 경우에 대한 예외 처리
    current_url = browser.current_url
    try:
        search_button = browser.find_element(By.XPATH, '//*[contains(@id, "card_tv_")]/div/div[2]/div[1]/div/div/a/h2')
        search_button.click()  # 검색 결과를 클릭하여 세부 정보 페이지로 이동
    except NoSuchElementException:
        continue  # 검색 결과가 없는 경우, 다음 반복으로 넘어감

    try:
        name = browser.find_element(By.XPATH, '//*[@id="original_header"]/div[2]/section/div[1]/h2/a').text
    except NoSuchElementException:
        name = 'none'

    try:
        plot = browser.find_element(By.XPATH, '//*[@id="original_header"]/div[2]/section/div[2]/div/p').text
    except NoSuchElementException:
        plot = 'none'

    try:
        genre = browser.find_element(By.XPATH, '//*[@id="original_header"]/div[2]/section/div[1]/div/span[2]').text
        if genre=='':
            genre='none'
    except NoSuchElementException:
        genre = 'none'

    try:
        img = browser.find_element(By.XPATH, '//*[@id="original_header"]/div[1]/div/div[1]/img').get_attribute('src')
        img = urljoin(current_url, img)
    except NoSuchElementException:
        img = 'none'

    try:
        button = browser.find_element(By.XPATH, '//*[@id="original_header"]/div[1]/div[2]/div/div[2]').get_attribute('href')
    except NoSuchElementException:
        button = 'none'

    name_ls.append(name)  # 각 리스트에 저장
    plot_ls.append(plot)
    genre_ls.append(genre)
    img_Is.append(img)
    button_ls.append(button)  # 현재 페이지 URL을 가져와서 저장
```

```python
import pandas as pd
pd.set_option('display.max_colwidth', -1)
sisa = {}
for col,val in zip(['name','plot','genre','img','buuton'],[name_ls,plot_ls,genre_ls,img_Is,button_ls]):
    sisa[col]=val

sisa=pd.DataFrame(sisa,columns=['name','plot','genre','img','buuton'])
```
